```python
hr_dataset_incomplete = hr_dataset.isnull()
hr_dataset_incomplete
```

```python
# create a new dataset where missing values are present
hr_dataset_missing = hr_dataset[hr_dataset.isnull().any(axis=1)]
hr_dataset_missing
```

```python
# to filter only those fields which are object in nature
hr_dataset_object = hr_dataset.select_dtypes(include=['object'])
hr_dataset_object
```

```python
# Create frequency table of Gender count along with percentage
gender_freq = hr_dataset['Gender'].value_counts()
gender_freq_percentage = hr_dataset['Gender'].value_counts(normalize=True) * 100
gender_freq, gender_freq_percentage
```

```python
# perform an exploratory data analysis on the dataset field Age to find out the mean, median, mode, min, max, standard deviation and variance
age_mean = hr_dataset['Age'].mean()
age_median = hr_dataset['Age'].median()
age_mode = hr_dataset['Age'].mode()[0]
age_min = hr_dataset['Age'].min()
age_max = hr_dataset['Age'].max()
age_std = hr_dataset['Age'].std()
age_var = hr_dataset['Age'].var()   
```

```python
hr_dataset['Age'].describe()
```

```python
# create a histogram to visualize the distribution of Age
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(hr_dataset['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# create a histogram to visualize the distribution of gender
plt.figure(figsize=(8, 5))
plt.bar(gender_freq.index, gender_freq.values, color='lightcoral', edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# calcuate average age of employees based on their gender
average_age_employees = hr_dataset.groupby('Gender')['Age'].mean()
average_age_employees
```

```python
# visualize average_age_employees
plt.figure(figsize=(8, 5))
plt.bar(average_age_employees.index, average_age_employees.values, color='lightgreen', edgecolor='black')
plt.title('Average Age of Employees by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Age')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
